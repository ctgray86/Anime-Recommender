Will create a collaborative filtering recomender engine and compare results versus a simple popularity recomender engine.

```python
from sklearn.model_selection import train_test_split
```

```python
data = graphlab.SFrame(ratings_new)
```

```python
train_data, test_data = data.random_split(0.8)
```

```python
model = graphlab.linear_regression.create(train_data, target='name', features = ['user_id', 'user_rating'])
```

```python
predictions = model.predict(test_data)
```

```python
results = model.evaluate(test_data)
```

```python
print model['training_rmse']
```

```python
popular = graphlab.popularity_recommender.create(model, user_id='user_id', item_id='name', target='user_rating')
```

```python
pop_rec = popular.recommend(users=range(1,6), k=10)
pop_rec.print_rows(num_rows=50)
```

```python
collabrative = graphlab.item_similarity_recommender.create(model, user_id='user_id', item_id='name',
                                                            target='user_rating', similarity_type='pearson')
```

```python
coll_rec = collabrative.recommend(users=range(1,6), k=10)
coll_rec.print_rows(num_rows=50)
```

```python
collabrative2 = graphlab.item_similarity_recommender.create(model, user_id='user_id', item_id='name',
                                                            target='user_rating', similarity_type='cosine')
```

```python
coll2_rec = collabrative2.recommend(users=range(1,6), k=10)
coll2_rec.print_rows(num_rows=50)
```

```python
coll_rec = collabrative.recommend(users=range(500,511), k=10)
coll_rec.print_rows(num_rows=100)
```
