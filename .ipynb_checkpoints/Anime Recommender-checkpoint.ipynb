Will create a collaborative filtering recomender engine and compare results versus a simple popularity recomender engine.

```python
actions = graphlab.SFrame(ratings_new)
```

```python
training_data, validation_data = graphlab.recommender.util.random_split_by_user(actions, 'user_id', 'name', 'user_rating')
model = graphlab.recommender.create(training_data, 'user_id', 'name', 'user_rating')
```

```python
testing_data = model.recommend()
```

```python
popular = graphlab.popularity_recommender.create(training_data, user_id='user_id', item_id='name', target='user_rating')
```

```python
pop_rec = popular.recommend(users=range(1,6), k=10)
pop_rec.print_rows(num_rows=50)
```

```python
pearson = graphlab.item_similarity_recommender.create(training_data, user_id='user_id', item_id='name',
                                                            target='user_rating', similarity_type='pearson')
```

```python
pearson_rec = pearson.recommend(users=range(1,6), k=10)
pearson_rec.print_rows(num_rows=50)
```

```python
cosine = graphlab.item_similarity_recommender.create(training_data, user_id='user_id', item_id='name',
                                                            target='user_rating', similarity_type='cosine')
```

```python
cosine_rec = cosine.recommend(users=range(1,6), k=10)
cosine_rec.print_rows(num_rows=50)
```

```python
recommender_performance = graphlab.compare(testing_data, [popular, pearson])
graphlab.show_comparison(recommender_performance,[popular, pearson])
```

```python
recommender_performance = graphlab.compare(testing_data, [popular, cosine])
graphlab.show_comparison(recommender_performance,[popular, cosine])
```

```python
recommender_performance = graphlab.compare(testing_data, [pearson, cosine])
graphlab.show_comparison(recommender_performance,[pearson, cosine])
```

```python
anime_list = pd.read_csv('../Datasets/animelist.csv')
```

```python
anime_list = anime_list.drop('Unnamed: 0', axis=1)
```

```python
genre_split2 = pd.concat([anime_list.genre.str.get_dummies(sep=", ")], axis=1)
```

```python
mas = MaxAbsScaler()
genre_split_fit = mas.fit_transform(genre_split2)
nn = NearestNeighbors(n_neighbors=11, algorithm='ball_tree').fit(genre_split_fit)
#After trying all the algorithms ball tree was the only one where the search wouldn't return itself everytime
distances, indices = nn.kneighbors(genre_split_fit)
```

```python
def obtain_animeindex(name):
    return anime_list[anime_list.name == name].index.tolist()[0]

def similar_anime(anime):
    obtained_animeindex = obtain_animeindex(anime)
    for id in indices[obtained_animeindex][1:]:
        print (anime_list.ix[id]['name'])
```
